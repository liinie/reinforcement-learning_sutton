```python
import numpy as np


def get_card(sum_card):
    card_value = np.random.choice([2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10, "ace"])
    if card_value == "ace":
        if sum_card + 11 <= 21:
            card_value = 11
            return int(card_value)
        else:
            card_value = 1
            return int(card_value)
    else:
        return int(card_value)


def play_game():
    reward = None
    done = False
    player_sum, dealer_sum = 0, 0
    # Initial 2 cards for dealer and player
    dealer_sum, player_sum = initial_round(dealer_sum, player_sum)
    # Player round
    player_sum, reward, done = player_round(dealer_sum, player_sum)
    if done:
        show_result(reward, player_sum, dealer_sum)
    else:
        assert done == False, "done is {}".format(done)
        # Dealer round
        dealer_sum, reward, done = dealer_round(dealer_sum, player_sum)
        if done:
            show_result(reward, player_sum, dealer_sum)
        else:
            print("something wrong")


def show_result(reward, player_sum, dealer_sum):
    if reward == 1:
        print("player wins!, player got card value {}, dealer got card value {}".format(player_sum, dealer_sum))
    elif reward == 0:
        print("draw!, player got card value {}, dealer got card value {}".format(player_sum, dealer_sum))
    else:
        assert reward == -1, "reward is {}".format(reward)
        print("player loses! player got card value {}, dealer got card value {}".format(player_sum, dealer_sum))


def dealer_round(dealer_sum, player_sum):
    while dealer_sum < 17:
        dealer_sum += get_card(dealer_sum)
    assert dealer_sum >= 17, "dealer_sum is {}".format(dealer_sum)
    if dealer_sum > 21:
        reward = 1
        done = True
    else:
        assert player_sum <= 21 and dealer_sum <= 21, "player sum is {} and dealer sum is {}".format(player_sum,
                                                                                                     dealer_sum)
        if player_sum > dealer_sum:
            reward = 1
            done = True
        elif player_sum < dealer_sum:
            reward = -1
            done = True
        elif player_sum == dealer_sum:
            reward = 0
            done = True
        else:
            raise ValueError("something wrong with the player sum or dealer sum")
    return dealer_sum, reward, done


def player_round(dealer_sum, player_sum):
    reward = None
    done = False
    if player_sum == 21:
        if dealer_sum == 21:
            reward = 0
            done = True
        else:
            reward = 1
            done = True
    else:
        while player_sum < 21:
            player_action = np.random.choice(['hit', 'stick'])
            if player_action == 'hit':
                player_sum += get_card(player_sum)
            else:
                break
        # assert player_sum >= 21, "player sum is {}".format(player_sum)
        if player_sum > 21:
            reward = -1
            done = True
        if player_sum == 21:
            # assert player_sum == 21, "player sum is {}".format(player_sum)
            reward = 1
            done = True

    return player_sum, reward, done


def initial_round(dealer_sum, player_sum):
    for _ in range(2):
        dealer_sum += get_card(dealer_sum)
        player_sum += get_card(player_sum)
    return dealer_sum, player_sum


if __name__ == '__main__':
    play_game()
```
